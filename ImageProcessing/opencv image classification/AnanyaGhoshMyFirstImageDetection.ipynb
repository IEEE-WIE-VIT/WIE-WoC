```python
import cv2  #pip install opencv-python
```

```python
import matplotlib.pyplot as plt #pip install matplotlib
```

```python
#https://gist.github.com/dkurt/54a8e8b51beb3bd3f770b79e56927bd7 #config
#https://github.com/opencv/opencv/wiki/TensorFlow-Object-Detection-API  #frozen model from latest MobileNet-SSD v3
config_file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model='frozen_inference_graph.pb'
```

```python
model=cv2.dnn_DetectionModel(frozen_model,config_file) #or 
#cv2.dnn.readNetFromTensorflow
```

```python
#https://github.com/pjreddie/darknet/blob/master/data/coco.names        #coco dataset, 80 labels
classLabels=[]
file_name='labels.txt'
with open(file_name,'rt') as fpt:
    classLabels=fpt.read().rstrip('\n').split('\n')
    #classLabels.append(fpt.read())
```

```python
print(classLabels)
```

```python
print(len(classLabels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)#255/2=127.5
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

#Read and Image

```python
img=cv2.imread('girl.png')
```

```python
plt.imshow(img)   #bgr
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidece, bbox= model.detect(img,confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale=3
font=cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    #cv2.rectangle(frame, (x,y),(x+w,y+h),(255,0,0),2)
    
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40),font, fontScale=font_scale,color=(0,255,0),thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cap = cv2.VideoCapture("Lab Experiment #3_ Types of Chemical Reactions. - Bing video and 5 more pages - Personal - Microsoft Edge 2021-03-07 22-03-30.mp4")

if not cap.isOpened():
    cap=cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot capture Video")
    
font_scale=3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    
    ClassIndex, confidece, bbox = model.detect(frame, confThreshold=0.55)
    
    print(ClassIndex)
    if(len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if(ClassInd<=80):
                cv2.rectangle(frame, boxes, color=(255,0,0), thickness=2)
                cv2.putText(frame, classLabels[ClassInd-1], (boxes[0]+10,boxes[0]+40) , font, fontScale=font_scale, color=(0,0,255), thickness=2)
    cv2.imshow('Object Detection', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
